```clojure
(ns dry-sunrise
  (:require
    [opencv4.utils :as u]
    [opencv4.gorilla :as g]
    [opencv4.core :refer :all]))
```

```clojure
(defn load-me[url]
 (-> 
   url
   (imread)
   (cvt-color! COLOR_RGB2GRAY)
   (u/resize-by 0.3)
   (u/>>>)))
```

```clojure
(load-me "resources/cat.jpg")
```

```clojure
(load-me "resources/lena.png")
```
